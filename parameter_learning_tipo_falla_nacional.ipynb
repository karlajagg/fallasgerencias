```
import pandas as pd
url= 'https://raw.githubusercontent.com/karlajagg/fallasgerencias/main/xtrainfallas-nacional5.csv'
data=pd.read_csv(url)
print (data)
```

```
from pgmpy.models import BayesianModel

model = BayesianModel([('AP', 'COND'),('AP', 'CHC'),('AP', 'AB'),('AP', 'EST'),\
                       ('AP', 'CIM') , ('COND','CECN'),  ('COND','PCI'),  \
                       ('PCI','CHH'),  ('CECN','EHG'),  ('CECN','LC'),  \
                       ('CECN','CHH'), ('AB','EHG'),  ('AB','FA'),  ('FA','EST'),\
                       ('FA','CHCH'),  ('CIM','EST'),  ('EST','LC'),\
                       ('CHCH', 'EHG'),('ST', 'AB'),('ST', 'CHCH'),('SEP','AP'),\
                       ('SEP', 'ST') , ('SEP','CHCH'),  ('SEP','Tipo falla'),  \
                       ('EHG','CIM'),  ('CHC','CECN'),  ('CHC','CHH'),\
                       ('CHC', 'AB') , ('CHC','EHG'),  ('CHC','ST'),  \
                       ('CHH','CIM'),  ('CHH','LC'),  ('CHH','ST')]) 
#El aprendizaje de parámetros es la tarea de estimar los valores de las 
#distribuciones de probabilidad condicional (CPD)
```

```
from pgmpy.estimators import ParameterEstimator
pe = ParameterEstimator(model, data)
#CONTADOR DE ESTADOS
print("\n", pe.state_counts('FA')) 
print("\n", pe.state_counts('LC')) 
print("\n", pe.state_counts('CIM')) 
print("\n", pe.state_counts('COND')) 
print("\n", pe.state_counts('CHH')) 
print("\n", pe.state_counts('EST'))
print("\n", pe.state_counts('EHG'))  
print("\n", pe.state_counts('PCI'))
print("\n", pe.state_counts('CHCH')) 
print("\n", pe.state_counts('ST'))
print("\n", pe.state_counts('AB'))
print("\n", pe.state_counts('CHC')) 
print("\n", pe.state_counts('AP'))
print("\n", pe.state_counts('CECN'))
print("\n", pe.state_counts('SEP'))
print("\n", pe.state_counts('Tipo falla'))
```

```
#Maximum Likelihood Estimation
from pgmpy.estimators import MaximumLikelihoodEstimator
mle = MaximumLikelihoodEstimator(model, data)
print(mle.estimate_cpd('FA')) 
print(mle.estimate_cpd('LC')) 
print(mle.estimate_cpd('CIM'))  
print(mle.estimate_cpd('COND')) 
print(mle.estimate_cpd('CHH')) 
print(mle.estimate_cpd('EST')) 
print(mle.estimate_cpd('EHG')) 
print(mle.estimate_cpd('PCI')) 
print(mle.estimate_cpd('CHCH')) 
print(mle.estimate_cpd('ST'))
print(mle.estimate_cpd('AB')) 
print(mle.estimate_cpd('CHC')) 
print(mle.estimate_cpd('AP'))
print(mle.estimate_cpd('CECN')) 
print(mle.estimate_cpd('SEP')) 
print(mle.estimate_cpd('Tipo falla')) 
```

```
# Calibrate all CPDs of `model` using MLE:
model.fit(data, estimator=MaximumLikelihoodEstimator)
```

```
from pgmpy.estimators import BayesianEstimator
# BayesianEstimator, too, can be used via the fit()-method
# Aqui imprime todas las CPDs
model.fit(data, estimator=BayesianEstimator, prior_type="BDeu", equivalent_sample_size=10) # default equivalent_sample_size=5
for cpd in model.get_cpds():
    print(cpd)
```
